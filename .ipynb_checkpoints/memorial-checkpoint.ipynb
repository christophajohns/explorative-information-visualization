# Visualising memorial data

## Setup

```python
# Imports
import pandas as pd
import seaborn as sns
import json
```

```python
# Read in memorial data
f = open("memorial.json",)
json_data = json.load(f)
json_data
```

```python
# View data for one person
json_data['hits']['hits'][0]['_source']
```

```python
# Process data
data = [
    {
        "id": int(person['_source']['id']),
        "name": f"{person['_source']['eesnimi']} {person['_source']['perenimi']}" if "eesnimi" in person["_source"].keys() and "eesnimi" in person["_source"].keys() else None,
        "year_of_birth": int(person["_source"]["sünd"]) if "sünd" in person["_source"].keys() else None,
        "year_of_death": int(person["_source"]["surm"]) if "surm" in person["_source"].keys() else None,
        "related_ids": [int(related_person_id) for related_person_id in person["_source"]["pereseosID"].split(" ") if related_person_id != person['_source']['id']] if "pereseosID" in person["_source"].keys() else []
    }
    for person in json_data["hits"]["hits"]
    if "_source" in person
]

data[:5]
```

## Scatter plot for year of birth and year of death

```python
# Create DataFrame
df = pd.DataFrame(data).sort_values(by=["year_of_birth"]).reset_index(drop=True)[["id", "name", "year_of_birth", "year_of_death"]]
df
```

```python
# Remove missing data
df = df.dropna()
df
```

```python
# Remove false data
df = df.loc[(df.year_of_birth > 1800) & (df.year_of_death > 1900)]
df
```

```python
# Export data as csv
df.to_csv('memorial.csv', index=False)
```

```python
# Create scatter plot
sns.set(rc={
    'axes.facecolor':'black',
    'figure.facecolor':'black',
    'axes.grid': False,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.spines.bottom': False,
    'axes.spines.left': False
})
plot = sns.scatterplot(x="year_of_death", y="year_of_birth", data=df, color="silver")
plot.set(xlabel=None, ylabel=None)
```

```python
# Save figure
plot.figure.savefig("memorial.png")
```

## Graph for relationships

```python
# Process data
nodes = [{
    "id": person["id"],
    "name": person["name"],
    "year_of_birth": person["year_of_birth"],
    "year_of_death": person["year_of_death"]
} for person in data]

links = [{
    "source": person["id"],
    "target": related_id
} for person in data for related_id in person["related_ids"]]

graph_data = {
    "nodes": nodes,
    "links": links
}

print("Example node: ", graph_data["nodes"][0])
print("Example link: ", graph_data["links"][0])
```

```python
# Remove links containing unknown IDs
df_nodes = pd.DataFrame(nodes)
df_links = pd.DataFrame(links)

df_links = df_links.join(df_nodes.set_index("id"), on=["target"], how="inner")[["source", "target"]]
df_links
```

```python
# Remove duplicate links (other direction)
df_links = df_links.sort_values(by=["source", "target"]).loc[df_links.source <= df_links.target]
df_links
```

```python
# Update graph data
graph_data["links"] = df_links.to_dict("records")

print("Example node: ", graph_data["nodes"][0])
print("Example link: ", graph_data["links"][0])
```

```python
# Save graph data as json
with open('graph.json', 'w+', encoding='utf-8') as f:
    json.dump(graph_data, f, indent=4, ensure_ascii=False)
```

```python
# Import
import numpy as np

# Construct smaller sample for test
sample_links = df_links.iloc[:10]
sample_nodes = df_nodes.loc[df_nodes.id.isin(sample_links.source) | df_nodes.id.isin(sample_links.target)]


graph_data_sample = {
    "nodes": sample_nodes.replace({np.NaN : None}).to_dict("records"),
    "links": sample_links.replace({np.NaN : None}).to_dict("records")
}
graph_data_sample
```

```python
# Save test set of graph data as json
with open('graph_sample.json', 'w+', encoding='utf-8') as f:
    json.dump(graph_data_sample, f, indent=4, ensure_ascii=False)
```
